```python
import sklearn as sk
import seaborn as sns
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.svm import SVC, LinearSVC
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from scipy import stats
```

```python
#to give the file path of the train and test data
raw_train_data = pd.read_excel('trainDataset.xls', sheet_name='Sheet1')
raw_test_data = pd.read_excel('testDatasetExample.xls', sheet_name='Sheet1')
```

```python
def dataPreprocessingClassification(df_raw_data):
    #df_raw_data = pd.read_excel('trainDataset.xls')
    
    df_raw_data = df_raw_data.drop(['RelapseFreeSurvival (outcome)', 'ID'], axis = 1)
    
    outlierCols = (df_raw_data.drop(['pCR (outcome)', 'ER', 'PgR', 'HER2', 'TrippleNegative', 'ChemoGrade', 'Proliferation', 'HistologyType', 'LNStatus', 'TumourStage'], axis = 1)).columns

    clean, target = cleanData(df_raw_data, True)
    
    return clean, target
    

def cleanData(df_raw, is_train):

    #checks for duplicate values
    print(df_raw.duplicated().sum())
    print(df_raw)
    
    #converts the 999 values to NaN values due to functionality to remove NaN values
    df_raw = df_raw.replace(999, np.nan)
    print(df_raw.isna().sum().sum())
    
    #replaces NaN values with the medium of the column
    for col in df_raw:
        df_raw[col] = df_raw[col].fillna(df_raw[col].mode()[0])
        
    print(df_raw.isna().sum().sum())
    
    #drops any remaining NaN values
    df_raw = df_raw.dropna()
    print(df_raw.isna().sum().sum())
    
    #removes outliers
    print(df_raw.shape)
    if is_train == True:
        df_raw = df_raw[(np.abs(stats.zscore(df_raw)) < 3).all(axis = 1)]
    
    #scales the data
    if is_train == True:
        target = df_raw['pCR (outcome)']
        learning = df_raw.drop('pCR (outcome)', axis = 1)
    else:
        df_raw.head()
        target = df_raw['ID']
        learning = df_raw.drop('ID', axis = 1)
    
    cols = list(learning.columns)
    scaler = StandardScaler()
    scaler.fit(learning)
    scaled = scaler.transform(learning)
    data = pd.DataFrame(scaled, columns=cols)
    return data, target
    
cleanTrain, target = dataPreprocessingClassification(raw_train_data)
#applies pca for optimal feature selection

lda = LinearDiscriminantAnalysis()
lda.fit(cleanTrain, target)
reduced = lda.transform(cleanTrain)
print(reduced.shape)
cleanTrain = pd.DataFrame(reduced)
cleanTest, ID = cleanData(raw_test_data, False)
cleanTest = lda.transform(cleanTest)
print(cleanTest.shape)
print(target)

```

```python
gs_SVC = GridSearchCV(SVC(),{
    'kernel' : ['linear', 'poly', 'rbf', 'sigmoid'], #default rbf
    'degree' : [2,3], #default = 3, only applys to kernel: poly
}, cv=10, return_train_score=False, scoring = 'accuracy')
gs_SVC.fit(cleanTrain.to_numpy(),target)

preds = gs_SVC.predict(cleanTest)
#pred2 = gs_SVC.predict(cleanTrain)
dict_data = {'ID': ID,
             'Predicted PCR': preds}
df_out = pd.DataFrame(dict_data)
df_out.to_excel('FinalTestPCR.xls', sheet_name='Classification', index=False)
#accuracy_score(target,pred2)
```
